```python
from faker import Faker
import pandas as pd
import random


fake = Faker("en_AU")
agents_cleaned_data = pd.read_csv("agents_cleaned.csv")

agents_cleaned_data['Email'] = agents_cleaned_data['Email'].apply(
    lambda x: x if pd.notnull(x) else fake.email()
)

agents_cleaned_data['Phone'] = agents_cleaned_data['Phone'].apply(
    lambda x: x if pd.notnull(x) else fake.phone_number()
)

agents_cleaned_data['Website'] = agents_cleaned_data['Website'].apply(
    lambda x: x if pd.notnull(x) else fake.url()
)

for col in ['Sec_business_name', 'Sec_business_address', 'Sec_business_suburb',
            'Sec_business_state']:
    agents_cleaned_data[col] = agents_cleaned_data[col].apply(
        lambda x: x if pd.notnull(x) else fake.company()
    )

agents_cleaned_data['Sec_business_country'] = agents_cleaned_data['Sec_business_country'].apply(
    lambda x: x if pd.notnull(x) else fake.country()
)


languages = ['English', 'Mandarin', 'Hindi', 'Arabic', 'Vietnamese', 'Spanish','Cantonese','French']
agents_cleaned_data['Language'] = agents_cleaned_data['Language'].apply(
    lambda x: x if pd.notnull(x) else random.choice(languages)
)


agents_cleaned_data['Google Rating'] = agents_cleaned_data['Google Rating'].apply(
    lambda x: x if pd.notnull(x) else round(random.uniform(2.5, 5.0), 1)
)


agents_cleaned_data['Number of Rating'] = agents_cleaned_data['Number of Rating'].apply(
    lambda x: x if pd.notnull(x) else random.randint(1, 500)
)


agents_cleaned_data['Linkedin URL'] = agents_cleaned_data['Linkedin URL'].apply(
    lambda x: x if pd.notnull(x) else f"https://linkedin.com/in/{fake.user_name()}"
)
agents_cleaned_data['Facebook'] = agents_cleaned_data['Facebook'].apply(
    lambda x: x if pd.notnull(x) else f"https://facebook.com/{fake.user_name()}"
)
agents_cleaned_data['Instagram'] = agents_cleaned_data['Instagram'].apply(
    lambda x: x if pd.notnull(x) else f"https://instagram.com/{fake.user_name()}"
)


agents_cleaned_data['Scheduling on the webiste (Y/N)'] = agents_cleaned_data['Scheduling on the webiste (Y/N)'].apply(
    lambda x: x if pd.notnull(x) else random.choice(['Y', 'N'])
)



practice_areas = ['Student Visa', 'Skilled Migration', 'Family Visa', 'Business Visa']
agents_cleaned_data['Practice Area'] = agents_cleaned_data['Practice Area'].apply(
    lambda x: x if pd.notnull(x) else random.choice(practice_areas)
)

agents_cleaned_data['Years of experience'] = agents_cleaned_data['Years of experience'].astype(int)
agents_cleaned_data['MARN'] = agents_cleaned_data['MARN'].astype(int)


agents_cleaned_data.to_csv("agents_filled.csv", index=False)
print("save 'agents_filled.csv'")
```
