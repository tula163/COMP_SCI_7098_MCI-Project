```python
import requests
from bs4 import BeautifulSoup
import json
res = requests.get('https://immi.homeaffairs.gov.au/visas/getting-a-visa/visa-listing#')
res.encoding = 'utf-8'

content = res.text
# print(content)
soup = BeautifulSoup(content,'lxml')
# print(soup)
```

```python
script_tag = soup.find(id='ctl00_PlaceHolderMain_PageSchemaHiddenField_Input')
if script_tag:
    json_data = script_tag["value"] 



data = json.loads(json_data)
content_list = data.get("content", [])

visa_list = []
for category in content_list:
    category_name = category["text"]  
    block_html = category["block"]  
    
    soup = BeautifulSoup(block_html, "lxml")
    
    links = soup.find_all("a")
    for link in links:
        visa_name = link.text.strip()
        visa_url = "https://immi.homeaffairs.gov.au" + link["href"]  
        
        visa_list.append({
            "category": category_name,
            "visa_name": visa_name,
            "visa_url": visa_url
        })

print(f"succeffuly {len(visa_list)} visa information！")
print(json.dumps(visa_list, indent=4, ensure_ascii=False))

with open("../server/australia_visas.json", "w", encoding="utf-8") as f:
    json.dump(visa_list, f, ensure_ascii=False, indent=4)
```

```python
with open("../server/australia_visas.json", "r", encoding="utf-8") as f:
    visa_list = json.load(f)

category_dict = {}

for visa in visa_list:
    category = visa["category"].strip()  
    if category not in category_dict:
        category_dict[category] = []  
    
    category_dict[category].append({
        "visa_name": visa["visa_name"],
        "visa_url": visa["visa_url"]
    })

with open("../server/grouped_visas.json", "w", encoding="utf-8") as f:
    json.dump(category_dict, f, ensure_ascii=False, indent=4)

print("Work Done！")
```
